```python
from scipy.stats import lognorm
```

Випадкова велиина $Y$ має логнормальний розподіл, якщо $Y = \exp \left\{ X \right\}$, де $X$ - випадкова величина з нормального розподілу.

```python
mu = 0
sigma = 1
```

Параметри $s, loc, scale$ відповідають логнормальному розподілу зі щільністю $$f(x) = \frac{1}{s (x - loc) \sqrt{2 \pi}} \exp \left\{ -\frac{1}{2 s^2} \log^2 \left( \frac{x - loc}{scale} \right) \right\}.$$

```python
s = sigma
loc = 0
scale = np.exp(mu)
```

Зобразіть графічно функцію щільності, функцію розподілу та нормовану гістограму $100000$ випадкових величин з параметром `bins=50` разом з функцією щільності рівномірного розподілу з параметрами:
1. $a = 0, b = 2$
2. $a = 0, b = 5$
3. $a = 0, b = 10$
4. $a = -2, b = 2$
5. $a = -5, b = 5$

Зробіть шкали на відповідних графіках однаковими для кращого порівняння. Наприклад, усі графіки щільності повинні мати однакові границі як по осі $x$, так і по осі $y$.

```python
# Місце для Вашого коду
from scipy.stats import uniform

a = [0, 0, 0, -2, -5]
b = [2, 5, 10, 2, 5]

plt.figure(figsize=(20,30))

counter = 0

for j in zip(a, b):
    x = np.arange(j[0], j[1] + 0.1, 0.1)
    
    counter += 1
    plt.subplot(5, 3, counter)
    plt.plot(x, uniform.pdf(x, *j))
    plt.xlim(j)
    plt.ylim([0, 1])
    plt.title("Probability density function")
    plt.ylabel("a={}, b={}".format(*j))
    
    counter += 1
    plt.subplot(5, 3, counter)
    plt.plot(x, uniform.cdf(x, *j))
    plt.xlim(j)
    plt.ylim([0, 1])
    plt.title("Cumulative distribution function")

    counter += 1
    plt.subplot(5, 3, counter)
    y = uniform.rvs(*j, size=100000)
    plt.hist(y, density=True, bins=50)
    plt.plot(x, uniform.pdf(x, *j))
    plt.xlim(j)
    plt.ylim([0, 1])
    plt.title("Cumulative distribution function\nand histogram of 100000 random elements")
```

```python
# Місце для Вашого коду
from scipy.stats import uniform

fig, ax = plt.subplots(5, 1, figsize=(8, 45))

def e_prob(a, n, i):
    pi = 0
    x = uniform.rvs(0, 2*a, size=n)
    y = uniform.rvs(0, 2*a, size=n)
    points = zip(x, y)
    
    circle = plt.Circle((a, a), 1, color='b', alpha=0.5)
    ax[i].add_artist(circle)
    
    for p in points:
        if (a - p[0])**2 + (a - p[1])**2 < a ** 2:
            pi += 1
            ax[i].scatter(*p, color='g')
        else:
            ax[i].scatter(*p, color='r')
            
    ax[i].set_xlim([0, 2 * a])
    ax[i].set_ylim([0, 2 * a])
    
    ax[i].set_title("n = {}, pi = {}".format(n, 4 * pi / n))
        
    return 4 * pi / n
    

a = 1
n = [10, 100, 500, 1000, 2000]
i = 0
for n in n:
    ep = e_prob(a, n, i)
    print(ep)
    i += 1
```
