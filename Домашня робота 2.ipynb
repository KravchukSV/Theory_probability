## 2 Герби в ряд

- $k = 2, n = 10, 50, 100, 500, 1000, 5000, 10000$
- $k = 3, n = 10, 50, 100, 500, 1000, 5000, 10000$

**Зауваження.** Нехай $k = 2, n = 10$ та маємо таку послідовність результатів підкидання монети: $\{ H, T, H, H, H, H, T, T, H, T \}$. Тоді матимемо $3$ випадки випадіння двох гербів підряд та емпіричну ймовірність такої події $1/3$.

```python
# Місце для Вашого коду
import numpy as np
from itertools import product

def emp_prob(k, n):
    a = ["H", "T"]
    m = 0
    replace = True
    A = list(np.random.choice(a, size=n, replace=replace))
    if k == 2:
        for i in range(n-1):
            if(A[i]==A[i+1] and A[i]=="H"):
                m+=1
        emp1 = (m/n)
    if k == 3:
        for i in range(n-2):
            if(A[i]==A[i+1] and A[i+1]==A[i+2] and A[i]=="H"):
                m+=1
        emp1 = (m/n)
    print("k = ", k, 'n = ', n)
    print("Емпірична ймовірність: {}".format(emp1))

def teor_prob(k, size):
    a = ["H", "T"]
    m = 0
    lst = list(product(a,repeat = size))
    if k == 2:
        for i in range(len(lst)):
            c = 0
            for j in range(len(lst[i])-1):
                if lst[i][j]==lst[i][j+1] and lst[i][j] == "H":
                    c+=1
            if c > 0:
                m+=1
    if k == 3:
        for i in range(len(lst)):
            c = 0
            for j in range(len(lst[i])-2):
                if lst[i][j]==lst[i][j+1] and lst[i][j+1]==lst[i][j+2] and lst[i][j] == "H":
                    c+=1
            if c > 0:
                m+=1
    print('k = ', k, ', n = ', size)
    print('Теортична ймовірність: ', m/len(lst))
    
    

n = 10
teor_prob(2, n)
for i in range(7):
    emp_prob(2, n)
    if i%2==0:
        n *= 5
    else:
        n *= 2
n = 10
print('\n')
teor_prob(3, n)
for i in range(7):
    emp_prob(3, n)
    if i%2==0:
        n *= 5
    else:
        n *= 2
```

## 3 Найбільше число
Реалізуйте функцію підрахунку точної (теоретичної) імовірності того, що максимальне число після $n$ кидків звичайного грального кубика дорівнює $x\_max$, для таких параметрів:
- $n = 3, x\_max = 6$
- $n = 2, x\_max = 3$
- $n = 3, x\_max = 5$

Тобто, якщо $X_i, i = 1, \ldots, n$, - число на $i$-му кидку кубика, а $X = \max\limits_{1 \leq i \leq n} X_i$, то потрібно знайти ймовірність $\textbf{P} \left( X = x\_max \right)$.

Підрахуйте також емпіричні ймовірності відповідних подій у випадку $10000$ симуляцій.

**Зауваження.** Якщо введені параметри не мають сенсу, то повертати $0$.
Тобто, якщо вводиться $x\_max < 1$ або $x\_max > 6$.

```python
# Місце для Вашого коду
from itertools import product
import numpy as np

A = [1, 2, 3, 4, 5, 6]

def teor_probs(n, x_max):
    if x_max > 6 or x_max<1:
        return 0 
    counter = 0
    lst = list(product(A, repeat = n))
    for i in range(len(lst)):
        y = 0
        for j in range(len(lst[i])):
            if lst[i][j]>y:
                y=lst[i][j]
        if y == x_max:
            counter += 1
    print('n = ', n, ', x_max = ', x_max)
    print('Tеоретична ймовірність: ', counter/len(lst))

def emp_probs(n, x_max):
    if x_max>6 or x_max<1:
        return 0
    replace = True
    siz = 10000
    counter = 0
    
    for l in range(siz):
        lst = list(np.random.choice(A, size=n, replace=replace))
        y = 0
        for i in range(len(lst)):
            if y < lst[i]:
                y = lst[i]
        if y == x_max:
            counter+=1
    print('n = ', n, ', x_max = ', x_max)
    print('Емпірична ймовірність: ', counter/siz
          )
    

teor_probs(3, 6)
emp_probs(3, 6)
teor_probs(2, 3)
emp_probs(2, 3)
teor_probs(3, 5)
emp_probs(3, 5)
```

## 4 Сума чисел
Реалізуйте функцію підрахунку точної (теоретичної) імовірності того, що сума чисел після $n$ кидків звичайного грального кубика дорівнює $s$, для таких параметрів:
- $n = 3, s = 13$
- $n = 2, s = 5$
- $n = 3, s = 7$

Тобто, якщо $X_i, i = 1, \ldots, n$, - число на $i$-му кидку кубика, а $X = \sum\limits_{i = 1}^{n} X_i$, то потрібно знайти ймовірність $\textbf{P} \left( X = s \right)$.

Підрахуйте також емпіричні ймовірності відповідних подій у випадку $10000$ симуляцій.

**Зауваження.** Якщо введені параметри не мають сенсу, то повертати $0$.
Наприклад, якщо при $n = 3$ вводиться $s < 3$ або $s > 18$.

```python
# Місце для Вашого коду
from itertools import product
import numpy as np

A = [1, 2, 3, 4, 5, 6]

def teor_probs(n, s):
    if s < n or s>n*6:
        return 0
    counter = 0
    lst = list(product(A, repeat = n))
    for i in range(len(lst)):
        suma = 0
        for j in range(len(lst[i])):
            suma += lst[i][j]
        if suma == s:
            counter += 1
    print('n = ', n, ', s = ', s)
    print('Tеоретична ймовірність: ', counter/len(lst))

def emp_probs(n, s):
    if s < n or s>n*6:
        return 0
    replace = True
    siz = 10000
    counter = 0
    
    for l in range(siz):
        lst = list(np.random.choice(A, size=n, replace=replace))
        suma = 0
        for i in range(len(lst)):
            suma += lst[i]
        if suma == s:
            counter+=1
    print('n = ', n, ', x_max = ', s)
    print('Емпірична ймовірність: ', counter/siz)
    

teor_probs(3, 13)
emp_probs(3, 13)
teor_probs(2, 5)
emp_probs(2, 5)
teor_probs(3, 7)
emp_probs(3, 7)
```
